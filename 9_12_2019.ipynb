```
// read two inputs (a1,a2) and print the output as "count of even palindrome
// test case 1 to 100


#include<stdio.h>
int ispalindrome(int n)
{
	int rev=0,buffer=n;
	while(n!=0)
	{
		rev=rev*10+(n%10);
		n/=10;
	}
	if (buffer==rev){
		return 1;
	}
	else{
		return 0;
	}
}
int main()
{
	int i,a1,a2,res,cnt=0;
	scanf("%d%d",&a1,&a2);
	for (i=a1;i<=a2;i++)
	{
		if(i%2==0)
		{
			res=ispalindrome(i);
			if(res==1){
				cnt++;
			}
		}
	}
	printf("%d",cnt);
	return 0;

}
```

```
// print yes or no if sum of individual factorial sum is same as input
// test cases 145=1!+4!+5!=145
              123=1!+2!+3!= 9
              


#include<stdio.h>
int factorial(int x)
{
	int i,fact=1;
	for(i=x;i>0;i--)
	{
		fact*=i;
	}
	return fact;
}
int main()
{
	int n,s=0,r,buffer;
	scanf("%d",&n);
	buffer=n;
	while (n!=0){
		r=n%10;
		s=s+factorial(r);
		n/=10;
	}
	if(s==buffer)
	{
		printf("yes");
	}
	else{
		printf("no");
	}
	return 0;
}
```

```
// read the two limits as input and print the count of perfect number in the given limit
// test cases 1 to 10 only 1 perfect number i.e 6 =1+2+3
#include<stdio.h>
int isperfect(int n)
{
	int i,sum=0;
	for(i=1;i<=(n/2);i++)
	{
		if (n%i==0)
		{
			sum=sum+i;
		}
	}
	if (sum==n)
	{
		printf("%d ",sum);
		return 1;
	}
	else 
	{
		return 0;
	}
}
int main()
{
	
	int a1,a2,i,res,perfectcnt=0;
    scanf("%d%d",&a1,&a2); 
    for (i=a1;i<=a2;i++)
    {
    	res=isperfect(i);
        if(res==1){
        	perfectcnt++;
		}
	}
	printf("number of perfect numbers = %d",perfectcnt);
	return 0;
}

```

```
// read the input and print the output as sum of each digits to power of digit count 
// test cases :-
// 121 -- 10(1^3+2^3+1^3)
//1221 -- 114(1^4+2^4+2^4+1^4)

#include<stdio.h>
#include<math.h>
int digitcount(int n)
{
	int cnt=0;
	while(n!=0)
	{
		cnt++;
		n=n/10;
	
	}
	return cnt;
}
int main()
{
	int n,r,s,dgtcnt,sum=0;
	scanf("%d",&n);
	dgtcnt=digitcount(n);
	while(n!=0){
		r=n%10;
	    s= pow(r,dgtcnt);
	    sum=sum+s;
	    n=n/10;
	}
	printf("%d",sum);
	return 0;
}
```

```
// read the input as two numbers and print the series of prime numbers between two limits
#include<stdio.h>
int isprime(int n)
{
	int i,flag=0;
	for(i=2;i<=(n/2);i++)
	{
		if (n%i==0){
			flag=1;
			break;
		}
	}
	if (flag==0)
	{
		return 1;
	}
	else{
		return 0;
	}
}
int main()
{
	int i,a1,a2,prime=0;
	scanf("%d%d",&a1,&a2);
	for (i=a1;i<=a2;i++)
	{
		prime=isprime(i);
		if (prime==1){
			printf("%d ",i);
		}
	}
	return 0;
 } 
```
